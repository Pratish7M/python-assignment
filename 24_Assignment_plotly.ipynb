```python
'''Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.'''

import seaborn as sns
import plotly.express as px

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create a scatter plot
fig = px.scatter(titanic, x='age', y='fare', title='Scatter Plot of Age vs. Fare',
                 labels={'age': 'Age', 'fare': 'Fare'},
                 color='class' ) # Optional: color by passenger class
                 

# Show the plot
fig.show()
```

```python
# Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Create a box plot for the total bill by day
fig = px.box(tips, x='day', y='total_bill', title='Box Plot of Total Bill by Day',
             labels={'day': 'Day', 'total_bill': 'Total Bill'},
             color='day')  # Optional: color by day

# Show the plot
fig.show()

```

```python
'''Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.'''

import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Create a box plot for the total bill by day
fig = px.box(tips, x='day', y='total_bill', title='Box Plot of Total Bill by Day',
             labels={'day': 'Day', 'total_bill': 'Total Bill'},
             color='day')  # Optional: color by day

# Show the plot
fig.show()
```

```python

'''Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.'''


import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Create a histogram for total_bill by sex
fig = px.histogram(tips, x='sex', y='total_bill', 
                   pattern_shape='smoker', color='day',
                   title='Histogram of Total Bill by Sex',
                   labels={'sex': 'Sex', 'total_bill': 'Total Bill', 'smoker': 'Smoker', 'day': 'Day'},
                   barmode='group')  # Optional: group bars by sex

# Show the plot
fig.show()
```

```python
'''Q5. What is Distplot? Using Plotly express, plot a distplot.'''

import plotly.express as px

# Load the iris dataset
iris = px.data.iris()

# Create a scatter matrix plot
fig = px.scatter_matrix(iris, 
                        dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                        color='species',
                        title='Scatter Matrix Plot of Iris Dataset',
                        labels={'species': 'Species'})

# Show the plot
fig.show()
```
