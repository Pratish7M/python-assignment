```python
'''Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.'''

import seaborn as sns


# Here are five common plots that can be created using the Seaborn library, along with their uses:

# 1. **Scatter Plot (`sns.scatterplot`)**:
# **Use**: Visualizes the relationship between two continuous variables. It is useful for identifying 
# correlations, trends, and outliers.


sns.scatterplot(x='total_bill', y='tip', data=tips)


# 2. **Box Plot (`sns.boxplot`)**:
# **Use**: Displays the distribution of a dataset based on a five-number summary 
# (minimum, first quartile, median, third quartile, and maximum). It helps in identifying 
# outliers and understanding the spread and skewness of the data.


sns.boxplot(x='day', y='total_bill', data=tips)



# 3. **Histogram (`sns.histplot`)**:
#**Use**: Shows the distribution of a single continuous variable by dividing the data 
# into bins and counting the number of observations in each bin. It is useful for understanding 
# the frequency distribution of the data.


sns.histplot(tips['total_bill'], bins=10)
  

# 4. **Heatmap (`sns.heatmap`)**:
# **Use**: Represents data in a matrix form with colors indicating the values. It is particularly 
# useful for visualizing the correlation matrix and highlighting relationships between variables.
 
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm')
 

# 5. **Pair Plot (`sns.pairplot`)**:
# **Use**: Plots pairwise relationships in a dataset, including scatter plots for numerical 
# variables and histograms or KDE plots for the diagonal. It is useful for exploring relationships 
# between multiple variables simultaneously.
  
sns.pairplot(tips)
  

# These plots provide various ways to visualize data and uncover insights, making Seaborn a  
# powerful tool for data analysis.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Box plot for pclass vs age
sns.boxplot(x='pclass', y='age', data=titanic, ax=axes[0])
axes[0].set_title('Box plot of Age by Pclass')

# Box plot for pclass vs fare
sns.boxplot(x='pclass', y='fare', data=titanic, ax=axes[1])
axes[1].set_title('Box plot of Fare by Pclass')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset('flights')

# Pivot the dataset to create a matrix
flights_pivot = flights.pivot(index='year', columns='month', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap='YlGnBu')

# Display the plot
plt.show()
```
